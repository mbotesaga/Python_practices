```python
# Task: Create a Calculator using Loops and Conditions

# Description:

# In this task, you will develop a simple calculator that can perform basic arithmetic operations: addition, subtraction, multiplication, and division. The calculator will repeatedly ask the user for input until the user chooses to exit. You will use loops and conditional statements to handle user input and execute the correct operations.

# Steps:

# 2. Input Operation: Prompt the user to choose an operation by entering a corresponding number (e.g., 1 for addition, 2 for subtraction, etc.).

done = False

while not done:
    
    # 1. Prompt the User: Display a menu of available operations to the user:

    # Addition

    # Subtraction

    # Multiplication

    # Division

    # Exit
    
    print("Choose an operation to perform: \n\n"
        "1. Addition \n"
        "2. Subtraction \n"
        "3. Multiplication \n"
        "4. Division \n"
        "5. Exit"
    )
    try:
        select = int(input("\n Select an operation form by entering a corresponding number 1, 2, 3, 4 or 5 to exit: "))

        if select == 5:
            print("\n Bye, see you again!")
            break

        if select in [1, 2, 3, 4]:
            # 3. Input Numbers: After the user selects an operation, prompt the user to enter two numbers on which the operation will be performed.

            # Enter the first number
            number1 = int(input("\nEnter the first number "))
            # Enter the first number
            number2 = int(input("\nEnter the second number "))

            # 4. Perform Operation: Use conditional statements (if, elif, else) to perform the selected operation based on the user's input.
            # 5. Result: Display the result of the operation to the user.

            # Addition: Add the two numbers.

            if select == 1:
                print("\nSum: {} + {} = {}\n".format(number1, number2, number1 + number2))

            # Subtraction: Subtract the second number from the first.

            elif select == 2:
                print("\nSubtract: {} - {} = {}\n".format(number1, number2, number1-number2))

            # Multiplication: Multiply the two numbers.

            elif select == 3:
                print("\nMult: {} * {} = {}\n".format(number1, number2, number1 * number2))

            # Division: Divide the first number by the second (ensure that the second number is not zero).

            elif select == 4:
                # 7. Error Handling: Handle cases like division by zero or invalid inputs for numbers or operations.
                if number2 == 0:
                    print("\nError: Division by zero is not allowed.\n")
                else:
                    print("\nDiv: {} / {} = {}\n".format(number1, number2, number1 / number2))


        else:
                print("There's no such choice.")
                
    except ValueError:
        print("\nInvalid input. Please enter a number.\n")
        
    # 6. Loop: After showing the result, ask the user if they want to perform another calculation. If yes, repeat the process; if no, exit the program.
    while not done:   
        anotherTask = input("\n Would you like to perform another task? type 'y' or 'n': ")
        if anotherTask == "n":
            print("Bye, see you again!")
            done = True
        elif anotherTask == 'y':
            break
        else:
            print("type 'y' or 'n'")
```

```python
import img2pdf
import os

def convert_image_to_pdf(image_path, pdf_path):
    try:
        # Check if the image file exists
        if not os.path.isfile(image_path):
            print(f"Error: The file {image_path} does not exist.")
            return

        # Open the image file and convert to PDF
        with open(image_path, "rb") as image_file:
            pdf_bytes = img2pdf.convert(image_file.read())

        # Write the PDF bytes to a new file
        with open(pdf_path, "wb") as pdf_file:
            pdf_file.write(pdf_bytes)

        print(f"Conversion successful! PDF saved as {pdf_path}")

    except Exception as e:
        print(f"An error occurred: {e}")

# Specify the input image and output PDF paths
input_image = "/home/hp/Desktop/img_to_pdf/profile.jpg"
output_pdf = "/home/hp/Desktop/img_to_pdf/profile.pdf"

# Call the function
convert_image_to_pdf(input_image, output_pdf)
```

```python
# Create a class BankAccount with the following properties and methods:
# Properties: owner, balance (default value is 0)
# Methods:
# deposit(amount): Adds the amount to the balance.
# withdraw(amount): Subtracts the amount from the balance if there are enough funds.
# display_balance(): Prints the account balance.

class bankAccount:
    def __init__(self, owner, balance = 0):
        self.owner = owner
        self.balance = balance

    def deposit(self, deposit_amount):
        # self.balance += deposit_amount
        self.balance = self.balance + deposit_amount
        print("Successfully deposited")

    def withdraw(self, withdraw_amount):
        if withdraw_amount <= self.balance:
            self.balance = self.balance - withdraw_amount
            print("Successfully withdrawn")
        else:
            print("not enough balance")
    
    def display_balance(self):
        print("Your current balance: ", self.balance)
        
#Create objects

bankAccount1 = bankAccount("James", 100)

bankAccount1.deposit(100)
bankAccount1.withdraw(50)
#DISPLAY
bankAccount1.display_balance()
```
